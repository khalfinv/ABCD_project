<font size="5"><b>Bivariate Analysis<font size="5"><b> 

I divided the analysis into 2 parts - numerical features and categorial features.

<font size = "4"><b> Numerical features analysis <font size = "4"><b>

<font size = "3"><b> Find the numerical features <font size = "3"><b>

```python
#select only numerical columns
num_col = [num_feat for num_feat in df.columns if df[num_feat].dtype!='object' and df[num_feat].dtype.name!='category']

print("There are {} numerical features\n".format(len(num_col)))
print("The numerical features are:\n", num_col)
```

<font size = "3"><b> View of numerical features <font size = "3"><b>

<font size = "3"><b> Distribution of numerical features <font size = "3"><b>

```python
#draw histograms in same grid
def draw_histograms(dataset, variables, n_rows, n_cols):
    fig=plt.figure(figsize = (20,20))
    for i, var_name in enumerate(variables):
        print(i)
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        dataset[var_name].hist(bins=10,ax=ax)
        ax.set_title(var_name)
    plt.show()

draw_histograms(df, num_col, 5, 4)
```

<font size = "4"><b> Categorial features analysis <font size = "4"><b>

<font size = "3"><b> Find the categorial features <font size = "3"><b>

```python
#select only categorial columns
catg_col = [num_feat for num_feat in df.columns if df[num_feat].dtype.name == 'category']
catg_col.append('label')


print("There are {} categorial features\n".format(len(catg_col)))
print("The numerical features are:\n", catg_col)
```

View of categorial features

```python
df[catg_col].head()
```

<font size = "3"><b> Categorial features distribution<font size = "3"><b>

```python
#draw boxplots in same grid
def draw_barh(dataset, variables, n_rows, n_cols):
    fig = plt.figure(figsize = (20,20))
    for i, var_name in enumerate(variables):
        ax = fig.add_subplot(n_rows,n_cols,i+1)
        ax = dataset[var_name].value_counts().head(10).plot.barh()
        ax.set_title(var_name)
    
draw_barh(df, catg_col, 4, 4)
```

```python
for feature in catg_col:
    percentage = df[feature].value_counts()/len(df)*100
    print("Feature {}: \n{}".format(feature,percentage))
```

<font size = "4"><b> Multivariate Analysis <font size = "4"><b>

<font size = "3"><b> Correlation heatmap <font size = "3"><b>

```python
plt.subplots(figsize=(15,15))
ax = sns.heatmap(df[num_col].corr(), 
        xticklabels=df[num_col].corr().columns,
        yticklabels=df[num_col].corr().columns,
            cmap='coolwarm',
           annot = True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```

```python
sns.set(style="ticks", color_codes=True)
g = sns.pairplot(df[num_col+['label']], hue="label")
```

<font size="6"><b>Data pre-processing <font size="6"><b>

<font size="5"><b>Outlires<font size="5"><b>

<font size="4"><b>Outlires in numerical features<font size="4"><b>

```python
#draw boxplots in same grid
def draw_boxplots(dataset, variables, n_rows, n_cols):
    fig=plt.figure(figsize = (30,40))
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        ax = sns.boxplot(dataset[var_name])
        ax.set_title(var_name)
        ax.xaxis.label.set_visible(False)
    plt.show()

draw_boxplots(df[num_col],num_col, 5, 4)
```

```python
from scipy import stats
df = df[(np.abs(stats.zscore(df[num_col])) < 3).all(axis=1)]
```
